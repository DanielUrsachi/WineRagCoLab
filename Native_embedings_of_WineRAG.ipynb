```python
import chromadb
client = chromadb.Client()
import json

def getBackground(prompt):

  backgroundCollection = client.get_or_create_collection("taste")

  backgroundCollection.add(
    documents = [json.dumps(item, indent=2) for item in background],
    ids = [str(wine["rank"]) for wine in background]
  )

  dbBackgroundResults = backgroundCollection.query(query_texts=[prompt],
                            n_results=3)
  print("getBackground output: ")
  print(dbBackgroundResults)
  return dbBackgroundResults
# test
# getBackground('light, for seafood dinner')
```

```python

def getReviews(prompt):
  reviewsCollection = client.get_or_create_collection("reviews")
  userId = 1
  documents = [comment["text"] for review in reviews if review["userId"] == userId for comment in review["comments"]]
  ids = [str(userId) + "-" + str(comment["commentId"]) for review in reviews if review["userId"] == 1 for comment in review["comments"]]
  # print(documents)
  # print(ids)
  reviewsCollection.add(documents=documents, ids=ids)

  getReviews = reviewsCollection.query(query_texts=[prompt],
                            n_results=2)
  print("getReviews output: ")
  print(getReviews)
  return getReviews

# test
# getReviews('light, for seafood dinner')
```

```python
from google.colab import userdata

from openai import OpenAI

import os
import copy
import re
import json

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

openAiClient = OpenAI(api_key=OPENAI_API_KEY)

if openAiClient.api_key is None:
    raise Exception('OPENAI_API_KEY not set')

def insert_params(string, **kwargs):
    pattern = r"{{(.*?)}}"
    matches = re.findall(pattern, string)
    for match in matches:
        replacement = kwargs.get(match.strip())
        if replacement is not None:
            string = string.replace("{{" + match + "}}", replacement)

    print("template result: " + string)
    return string

def completion(template, params):
    instance = copy.deepcopy(template)

    for item in instance['messages']:
        item['content'] = insert_params(item['content'], **params)

    return openAiClient.chat.completions.create(model='gpt-3.5-turbo', **instance)
```

```python
template = {
  "temperature": 0,
  "messages": [
        {
          "role": "system",
          "content": """You are an AI assistant trying suggest the best wines (choose 3 to 6 wines) that fits client prompt. Take into consideration the wine backgorund reviews, and user previous reviews"""
        },
        {
          "role": "user",
          "content": """backround of wines reviews:{{background}}. user previous reviews: {{reviews}}"""
        }
    ],
}

# response = completion(template, {}).choices[0].message
# print("print(preconditions.content) " +preconditions.content)

while True:
    prompt = input('> ')
    if prompt == 'exit':
        break

    backgroundVal = "".join(getBackground(prompt)["documents"][0])
    reviewsVals = "".join(getReviews(prompt)["documents"][0])

    template['messages'].append({'role': 'user', 'content': prompt})
    message = completion(template, {'request': prompt, 'background': backgroundVal, 'reviews': reviewsVals}).choices[0].message

    print(message.content)
    # update conversation history
    template['messages'].append({'role': message.role, 'content': message.content})

    # prompts:
    # 'light, for seafood dinner'
```
